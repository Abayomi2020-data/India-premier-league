```python
balls = df_league.groupby("batsman")["ball"].count().reset_index()
```

```python
balls.head()
```

```python
runs = df_league.groupby("batsman")["batsman_runs"].sum().reset_index()
```

```python
runs.head()
```

```python
four = df_league[df_league["batsman_runs"]==4]
four.head()
```

```python
runs_4 = four.groupby("batsman")["batsman_runs"].count().reset_index()
```

```python
runs_4.head()
```

```python
runs_4.columns = ["batsman", "4s"]
runs_4.head()
```

```python
six = df_league[df_league["batsman_runs"]==6]
six.head()
```

```python
six = six.groupby("batsman")["batsman_runs"].count().reset_index()
```

```python
six.columns = ["batsman", "6s"]
six.head()
```

```python
player = pd.concat([runs, balls.iloc[:,1], runs_4.iloc[:,1],six.iloc[:,1]], axis=1)
player
```

```python
player.fillna(0, inplace=True)
```

```python
player.head()
```

```python
player["strike_rate"] = (player["batsman_runs"]/player["ball"])*100
```

```python
player.head()
```

```python
grp = df_league.groupby(["match_id", "batsman", "batting_team"])["batsman_runs"].sum().reset_index()
```

```python
grp.head()
```

```python
max = grp.groupby("batsman")["batsman_runs"].max().reset_index()
max.columns = ["batsman", "max_runs"]
max.head()
```

```python
player2 = pd.concat([player, max.iloc[:,1]], axis=1)
player2.head()
```
